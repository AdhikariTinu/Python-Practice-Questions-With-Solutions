## Patern Questions (Using for loops)

Right angled Triangle

```python
for i in range(1,6):
    print("*"*i)
```

```python
for row in range(1,6):
    for column in range(row):
        print("*",end="")
        column +=1
    row += 1
    print()
```

Inverted Triangle

```python
for rows in range(5,0,-1):
    for column in range(rows):
        print("*",end="")
        column -=1
    print()
    rows -=1
    
```

Number Pattern

```python
for row in range(2,7):
    for column in range(1,row):
        print(column,end="")
        column +=1
    row += 1
    print()
```

Inverted Number Pattern

```python
for row in range(6,0,-1):
    for column in range(1,row):
        print(column,end="")
        column-=1
    print()
    row-=1
```

Pyramid Pattern

```python
n = 5
for i in range(1, n + 1):
    for j in range(n - i):
        print(" ", end="")

    for k in range(1, 2*i):
        print("*", end="")
    print()
```

Alphabet Pattern

```python
ord("A")
```

```python
for i in range(1, 6):
    for j in range(i):
        print(chr(65 + j), end="")
    print()
```

Right aligned Triangle

```python
rows = 5
for i in range(1, rows + 1):
    for j in range(1, rows - i + 1):
        print(" ", end="")
        
    for k in range(1, i + 1):
        print("*", end="")
    print()
```

Hollow Square

```python
a = 5
for i in range(a):
    for j in range(a):
        if i == 0 or i == a - 1 or j == 0 or j == a - 1:
            print('*', end='')
        else:
            print(' ', end='')
    print()
```

Diamond Pattern

```python
rows = 3
for i in range(rows):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))

for i in range(rows - 2, -1, -1):
    print(" " * (rows - i - 1) + "*" * (2 * i + 1))
```

Pascal's Triangle

```python
n = 5
for i in range(n):
    print(" " * (n - i), end="")
    value = 1  
    for j in range(i + 1):
        print(value, end=" ")
        value = value * (i - j) // (j + 1)
    print()
```
